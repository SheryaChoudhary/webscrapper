```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Create an URL object
url = 'https://opentender.eu/it/dashboards/transparency'
# Create object page
page = requests.get(url)
```

```python
page
```

```python
# parser-lxml = Change html to Python friendly format
# Obtain page's information
soup = BeautifulSoup(page.text, 'lxml')
soup
```

```python
# Obtain information from tag <table>
table1 = soup.find('table', id='table-top')
table1
```

```python
# Obtain every title of columns with tag <th>
headers = []
for i in table1.find_all('td'):
    title = i.text
    headers.append(title)
```

```python
print(headers)
```

```python
# Removing newline character from stringusing loop
data = []
for sub in headers:
    data.append(sub.replace("\n", ""))
 
# printing result
print("List after newline character removal : " + str(data))
```
